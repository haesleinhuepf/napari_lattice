Importing libraries and loading the Dataset

```python
from skimage import io
import pyclesperanto_prototype as cle

location = 'D:/Uni/Hackathon/neotosis#/'
image = io.imread(location + 'netosis_ch2.tif')
```

Taking a look at the data

```python
cle.imshow(image)
```

Deskewing the image and showing the result

```python
deskewed = cle.deskew_y(image, angle_in_degrees=30, voxel_size_x=0.145, voxel_size_y=0.145, voxel_size_z=0.3)
cle.imshow(deskewed)
```
